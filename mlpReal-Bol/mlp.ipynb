Los datos de investing.com vienen ordenados del mas nuevo al mas viejo. Para que la red neuronal entienda el flujo del tiempo, debemos invertir esto y limpiar un poco las columnas 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir la ruta y Cargar el dataset
ruta_csv = r'D:\2026\IA-2026-DATASETS\BRL_BOB Historical Data.csv'

# ¡ESTA ES LA LÍNEA QUE FALTABA! Leemos el archivo y lo guardamos en 'df'
df = pd.read_csv(ruta_csv)

# 2. Nos quedamos solo con las columnas que importan: Fecha y Precio de Cierre
df = df[['Date', 'Price']]

# 3. Convertir la columna 'Date' a formato de fecha de Pandas
df['Date'] = pd.to_datetime(df['Date'])

# 4. Ordenar los datos cronológicamente (del más antiguo al más nuevo)
# Esto es CRUCIAL porque Investing.com los entrega al revés
df = df.sort_values('Date').reset_index(drop=True)

# 5. Asegurarnos de que el 'Price' sea numérico (a veces viene como texto con comas)
if df['Price'].dtype == 'object':
    df['Price'] = df['Price'].str.replace(',', '').astype(float)

# 6. Ver los primeros datos y graficar para comprobar
print("Tamaño del dataset:", df.shape)
print(df.head())

# Graficar la serie temporal
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Price'], color='blue')
plt.title('Cotización Histórica Real Brasileño a Boliviano (2015 - Actualidad)')
plt.xlabel('Fecha')
plt.ylabel('Precio (BOB)')
plt.grid(True)
plt.show()
```
